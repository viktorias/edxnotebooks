Based on these I'm going to separate these into 5 sections and find some average values.

```python
idx_tr1_begin = 3404
idx_tr1_end = 3436
display(time[idx_tr1_begin], flux[idx_tr1_begin])
display(time[idx_tr1_end], flux[idx_tr1_end])
```

```python
idx_tr2_begin = 14277
idx_tr2_end = 14308
display(time[idx_tr2_begin], flux[idx_tr2_begin])
display(time[idx_tr2_end], flux[idx_tr2_end])
```

```python
time = np.array(time)
flux = np.array(flux)

time_parts = np.split(time, [idx_tr1_begin, idx_tr1_end+1, idx_tr2_begin, idx_tr2_end+1])
flux_parts = np.split(flux, [idx_tr1_begin, idx_tr1_end+1, idx_tr2_begin, idx_tr2_end+1])
```

```python
avg_normal1 = np.mean(flux_parts[0])
avg_tr1 = np.mean(flux_parts[1])
avg_normal2 = np.mean(flux_parts[2])
avg_tr2 = np.mean(flux_parts[3])
avg_normal3 = np.mean(flux_parts[4])
display(avg_normal1, avg_tr1, avg_normal2, avg_tr2, avg_normal3)
```

```python
avg_normal = np.mean([avg_normal1, avg_normal2, avg_normal3])
avg_tr = np.mean([avg_tr1, avg_tr2])
display(avg_normal, avg_tr)
```

```python
flux_change = avg_normal - avg_tr
flux_change
```

According to JohnVSavage's Tyr analysis Tyr's radius is $6.26 \times 10^{8} m$.<br />
Based on the we can calculate the planet's radius.

```python
r_s, Delta_B, B = symbols("r_s, Delta_B, B", real=True)
r_p = r_s * sqrt(Delta_B/B)
r_p
```

```python
rp_val = r_p.subs([(Delta_B, flux_change), (B, avg_normal), (r_s, 6.26e8)])/1000
rp_val
```

The radius of Earth is 6378.1km.

```python
rp_val/6378.1
```

So it is around 3.5 bigger than Earth. The radius of Neptune is 24,622 km, so this planet is as big as Neptune.<br />
Let's calculate the planet's distance from the star:

```python
G, m_st, P, A = symbols("G m_st P A",real = True)
r_po = (G*m_st*P**2/(4*pi**2))**Rational(1,3)
r_po
```

```python
time[idx_tr2_end]-time[idx_tr1_end]
```

```python
P_val = (time[idx_tr2_begin]-time[idx_tr1_begin]+time[idx_tr2_end]-time[idx_tr1_end])*3600/2
P_val
```

According to JohnVSavage's Tyr analysis Tyr's mass is $1.75 \times 10^{30}$.

```python
r_po_val = r_po.subs([(P, P_val), (G, 6.67e-11), (m_st, 1.75e30)]).evalf()
r_po_val
```
