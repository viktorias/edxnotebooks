```python
from IPython.display import display
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Name,type,RA (deg),Distance (pc),Flux (W/m2),Temp (K),Vrad (m/s)
#List RA horizontal type vertical
x = []
y = []
f=open("data/stars_named_clue4_revised.csv")
i = 0
for row in csv.reader(f):
    if (i==0):
        i+=1
        continue
    #print(row)
    x.append(float(row[2]))
    y.append(float(row[1]))
plt.plot(x,y, 'ro')
plt.xlabel('RA (deg)')
plt.ylabel('Type')
plt.axis([-180, 180, 0, 3])
plt.show()
```

Odin is a type 0 star.

```python
plt.plot(x,y, 'ro')
plt.xlabel('RA (deg)')
plt.ylabel('Type')
plt.axis([-25, 25, 0, 3])
plt.show()
```

Type 1 stars are from -20 to 20 RA.

```python
#Name,type,RA (deg),Distance (pc),Flux (W/m2),Temp (K),Vrad (m/s)
#List RA horizontal type vertical
ra_rad1 = []
dist1 = []
flux1 = []
temp1 = []
rad_vel1 = []
name1 = []

ra_rad2 = []
dist2 = []
flux2 = []
temp2 = []
rad_vel2 = []
name2 = []

odin = []
loki = []

f=open("data/stars_named_clue4_revised.csv")
i = 0
for row in csv.reader(f):
    if (i > 0 and row[0] == 'Loki'):
        loki = [float(row[2])*np.pi/180, float(row[3]), \
                float(row[4]), float(row[5]), float(row[6]), row[0]]
    elif (i > 0 and row[0] == 'Odin'):
        #This is the data for Odin: Odin,0,0,57.3,5E-05,,0
        #Temp unknown => let it be -1
        odin = [float(row[2])*np.pi/180, float(row[3]), \
                float(row[4]), -1.0, float(row[6]), row[0]]
    elif (i > 0 and int(row[1]) == 1):
        ra_rad1.append(float(row[2])*np.pi/180) #in radians!!!
        dist1.append(float(row[3]))
        flux1.append(float(row[4]))
        temp1.append(float(row[5]))
        rad_vel1.append(float(row[6]))
        name1.append(row[0])
    elif (i > 0 and int(row[1]) == 2):
        ra_rad2.append(float(row[2])*np.pi/180) #in radians!!!
        dist2.append(float(row[3]))
        flux2.append(float(row[4]))
        temp2.append(float(row[5]))
        rad_vel2.append(float(row[6]))
        name2.append(row[0])
    elif (i==0):
        i+=1
```

```python
ra_rad1 = np.array(ra_rad1)
dist1 = np.array(dist1)
flux1 = np.array(flux1)
temp1 = np.array(temp1)
rad_vel1 = np.array(rad_vel1)

ra_rad2 = np.array(ra_rad2)
dist2 = np.array(dist2)
flux2 = np.array(flux2)
temp2 = np.array(temp2)
rad_vel2 = np.array(rad_vel2)
```

#### Distance plot

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)

plt.plot(ra_rad1, dist1, 'b*', label='Type 1')
plt.plot(ra_rad2, dist2, 'go', label='Type 2')
plt.plot(odin[0], odin[1], 'ks', label='Odin')
plt.plot(loki[0], loki[1], 'rs', label='Loki')
plt.legend()
plt.show()
```

The closest stars:

```python
min_dist = np.amin(dist1)
indices = np.where( dist1 == min_dist )
display("-from the type 1 stars:" + name1[indices[0][0]]+" "+str(min_dist)+"pc away")
min_dist = np.amin(dist2)
indices = np.where( dist2 == min_dist )
display("-from the type 2 stars:" + name2[indices[0][0]]+" "+str(min_dist)+"pc away")
display("Odin:" + str(odin[1])+"pc away")
display("Loki:" + str(loki[1])+"pc away")
```

Loki is the closest star among them :(.

#### Temperature plots

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.title('Temperature of the Type 1 stars (K)\n')
c = plt.scatter(ra_rad1, dist1, c=temp1, cmap=plt.cm.autumn, marker='o', s=40, label='Type 1')
c.set_alpha(0.75)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.title('Temperature of the Type 2 stars (K)\n')
plt.scatter(ra_rad2, dist2, c=temp2, cmap=plt.cm.autumn, \
            marker='o', s=40, label='Type 2', vmin=1800, vmax=8000, alpha=0.75)
plt.scatter(loki[0], loki[1], c=[loki[3]], cmap=plt.cm.autumn, \
            marker='s', s=40, label='Loki', vmin=1800, vmax=8000, alpha=0.75)
plt.colorbar()
#plt.legend()
plt.show()
```

It seems that the type 1 stars are much more hot than the type 2 stars. Loki is one of coolest stara here. (Odin's temperature is not in the data.)<br />
The minimum temperatures are:

```python
min_temp = np.amin(temp1)
indices = np.where( temp1 == min_temp )
display("-from the type 1 stars:" + name1[indices[0][0]]+" "+str(min_temp)+"K")
min_temp = np.amin(temp2)
indices = np.where( temp2 == min_temp )
display("-from the type 2 stars:" + name2[indices[0][0]]+" "+str(min_temp)+"K")
display("Loki:" + str(loki[3])+"K")
```

The maximum temperatures are:

```python
max_temp = np.amax(temp1)
indices = np.where( temp1 == max_temp )
display("-from the type 1 stars:" + name1[indices[0][0]]+" "+str(max_temp)+"K")
max_temp = np.amax(temp2)
indices = np.where( temp2 == max_temp )
display("-from the type 2 stars:" + name2[indices[0][0]]+" "+str(max_temp)+"K")
```

#### Radial velocity plots

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.title('Radial velocity of the type 1 stars\n')
plt.scatter(ra_rad1, dist1, c=rad_vel1, cmap=plt.cm.RdYlGn, \
            marker='o', s=40, label='Type 1', alpha=0.75)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.title('Radial velocity of the type 2 stars\n')
plt.scatter(ra_rad2, dist2, c=rad_vel2, cmap=plt.cm.RdYlGn, \
            marker='o', s=40, label='Type 2', alpha=0.75)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)
plt.title('Radial velocity of the type 1 & type 2 stars\n')
plt.scatter(ra_rad1, dist1, c=rad_vel1, cmap=plt.cm.RdYlGn, \
            marker='s', s=40, label='Type 1', alpha=0.75, vmin=-3910, vmax=3060)
plt.scatter(ra_rad2, dist2, c=rad_vel2, cmap=plt.cm.RdYlGn, \
            marker='o', s=40, label='Type 2', alpha=0.75, vmin=-3910, vmax=3060)
plt.colorbar()
plt.legend()
plt.show()
```
