```python
from IPython.display import display

from sympy.interactive import printing
printing.init_printing()
from __future__ import division
import sympy as sym
from sympy import *

import csv
import numpy as np
from scipy.optimize import leastsq
import matplotlib.pyplot as plt
%matplotlib inline
```

So the file contains Time (days), Radial Velocity (m/s). Radial velocity precisions of better than 3 m/s on Mog.<br />
So the goal is to fit a sin wave to the function in order to figure out the amplitudo and the period. And do some calcualtions based on that data.

```python
time = []
rad_vel = []
sum_val = 0.0
max_val = -1000.0
min_val = 1000.0
val = 0

f=open("data/Tyr_rv_4_revised.csv")
i = 0
for row in csv.reader(f):
    if (i > 0):
        time.append(float(row[0]))
        val = float(row[1])
        rad_vel.append(val)
        sum_val += val
    i+=1
```

```python
plt.figure(figsize=(20,10))
plt.plot(time, rad_vel, 'bo')
plt.xlabel('Time (days)')
plt.ylabel('Radial Velocity (m/s)')
plt.show()
```

```python
i-=1
avg_vel = sum_val/i
display("Avg: " + str(avg_vel))
```

```python
time = np.array(time)
rad_vel = np.array(rad_vel)
y = rad_vel - avg_vel
#display(y)
y_min = np.absolute(np.amin(y))
y_max = np.absolute(np.amax(y))
A = np.maximum(y_min, y_max)
#display(A)
y = y/A
y = np.arcsin(y)
```

```python
plt.plot(y, time, 'bo')
coeffs = np.polyfit(y, time, 1)
x2 = np.linspace(-1, 2)
y2 = coeffs[0]*x2+coeffs[1]
plt.plot(x2, y2, 'g')
plt.show()
```

```python
x_time = np.linspace(0, 180)
f = 1/coeffs[0]
phi = coeffs[1]
y3 = A*np.sin(f*(x_time-phi)) + avg_vel
plt.figure(figsize=(20,10))
plt.plot(time, rad_vel, 'bo')
plt.plot(x_time, y3, 'r')
plt.xlabel('Time (days)')
plt.ylabel('Radial Velocity (m/s)')
plt.show()
```

```python
func_wanted = lambda c_opt: c_opt[0]*np.sin(c_opt[1]*time-c_opt[2]) + c_opt[3] - rad_vel #original x original y
coeffs2 = leastsq(func_wanted, [A, f, phi*f, avg_vel])[0] #const what we have found
```

```python
display(coeffs2[1], coeffs2[0])
y3 = coeffs2[0]*np.sin(coeffs2[1]*x_time-coeffs2[2]) + coeffs2[3]
plt.figure(figsize=(20,10))
#plt.plot(time, rad_vel, 'bo')
plt.errorbar(time, rad_vel, yerr=3, fmt='o')
plt.plot(x_time, y3, 'r')
plt.xlabel('Time (days)')
plt.ylabel('Radial Velocity (m/s)')
plt.show()
```

According to JohnVSavage's Tyr analysis Tyr's mass is $1.75 \times 10^{30}$.<br />
If there is a planet around Tyr and the orbit is edge-on we can calculate the  radius of the reflex motion circle.

```python
G, m_st, P, A = symbols("G m_st P A",real = True)
r_po = (G*m_st*P**2/(4*pi**2))**Rational(1,3)
r_po
```

```python
r_po_val = r_po.subs([(P, P_val), (G, 6.67e-11), (m_st, 1.75e30)]).evalf()
r_po_val
```

The Sun-Earth distance is 149597870700 m.

```python
r_po_val/149597870700
```

So if the planet is edge-on orbit then it radius of the reflex motion circle is around 30% of the Sun-Earth distance that is as close as Mercury.

```python
m_sini = P*v_s*m_st/(2*pi*r_po)
m_sini
```
